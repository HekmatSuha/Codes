```python
#Task 1
from tkinter import *
from tkinter import ttk

# Function to change the background color of the window
def change_bg_color(color):
    root.configure(bg=color)

root = Tk()
root.title("Change Background Color")
root.geometry("300x200")

# Create buttons with different colors
btn_red = ttk.Button(root, text="Red", command=lambda: change_bg_color("red"))
btn_green = ttk.Button(root, text="Green", command=lambda: change_bg_color("green"))
btn_blue = ttk.Button(root, text="Blue", command=lambda: change_bg_color("blue"))

# Position the buttons
btn_red.pack(side=LEFT, padx=10, pady=20)
btn_green.pack(side=LEFT, padx=10, pady=20)
btn_blue.pack(side=LEFT, padx=10, pady=20)

root.mainloop()
```

```python
#Task 4 Translation
from tkinter import *
from tkinter import ttk

# Predefined dictionary for translation
translation_dict = {
    "hello": "привет",
    "world": "мир",
    "cat": "кот",
    "dog": "собака",
    "good": "хорошо",
    "bad": "плохо",
    "morning": "утро",
    "evening": "вечер"
}

def translate():
    word = english_word.get().lower()
    translated_word = translation_dict.get(word, "Translation not found")
    translation_box.config(state=NORMAL)
    translation_box.delete(1.0, END)
    translation_box.insert(END, translated_word)
    translation_box.config(state=DISABLED)

def clear_translation():
    english_word_entry.delete(0, END)
    translation_box.config(state=NORMAL)
    translation_box.delete(1.0, END)
    translation_box.config(state=DISABLED)

root = Tk()
root.title("English to Russian Translator")
root.geometry("400x300")

# English word input
english_word_label = Label(root, text="Enter an English word:", font=("Arial", 14))
english_word_label.pack(pady=10)

english_word = StringVar()
english_word_entry = ttk.Entry(root, textvariable=english_word, font=("Arial", 14))
english_word_entry.pack(pady=5)

# Translate button
translate_button = ttk.Button(root, text="Translate", command=translate)
translate_button.pack(pady=10)

# Clear button
clear_button = ttk.Button(root, text="Clear", command=clear_translation)
clear_button.pack(pady=10)

# Translation result in a text box
translation_label = Label(root, text="Translation in Russian:", font=("Arial", 14))
translation_label.pack(pady=10)

translation_box = Text(root, height=1, width=30, font=("Arial", 14), state=DISABLED)
translation_box.pack(pady=5)

root.mainloop()



```

```python
#Conversion Tasks

import tkinter as tk

def cm_to_m(distance_cm):
    return distance_cm / 100

def m_to_cm(distance_m):
    return distance_m * 100

def convert_cm_to_m():
    try:
        distance_cm = float(cm_entry.get())
        meters = cm_to_m(distance_cm)
        cm_result_label.config(text=f"{distance_cm} cm is {meters:.2f} meters.")
    except ValueError:
        cm_result_label.config(text="Please enter a valid number.")

def convert_m_to_cm():
    try:
        distance_m = float(m_entry.get())
        centimeters = m_to_cm(distance_m)
        m_result_label.config(text=f"{distance_m} meters is {centimeters:.2f} cm.")
    except ValueError:
        m_result_label.config(text="Please enter a valid number.")

# Create the main window
window = tk.Tk()
window.title("Centimeter to Meter and Meter to Centimeter Converter")

# CM to M conversion
cm_label = tk.Label(window, text="Enter distance in centimeters:")
cm_label.pack()
cm_entry = tk.Entry(window)
cm_entry.pack()
cm_btn = tk.Button(window, text="Convert cm to meters", command=convert_cm_to_m)
cm_btn.pack()
cm_result_label = tk.Label(window, text="")
cm_result_label.pack()

# M to CM conversion
m_label = tk.Label(window, text="Enter distance in meters:")
m_label.pack()
m_entry = tk.Entry(window)
m_entry.pack()
m_btn = tk.Button(window, text="Convert meters to cm", command=convert_m_to_cm)
m_btn.pack()
m_result_label = tk.Label(window, text="")
m_result_label.pack()

# Run the Tkinter event loop
window.mainloop()



```

```python
# kilograms to tons 
import tkinter as tk

def kg_to_tons(weight_kg):
    return weight_kg / 1000

def tons_to_kg(weight_tons):
    return weight_tons * 1000

def convert_kg_to_tons():
    try:
        weight_kg = float(kg_entry.get())
        tons = kg_to_tons(weight_kg)
        kg_result_label.config(text=f"{weight_kg} kg is {tons:.3f} tons.")
    except ValueError:
        kg_result_label.config(text="Please enter a valid number.")

def convert_tons_to_kg():
    try:
        weight_tons = float(tons_entry.get())
        kilograms = tons_to_kg(weight_tons)
        tons_result_label.config(text=f"{weight_tons} tons is {kilograms:.3f} kg.")
    except ValueError:
        tons_result_label.config(text="Please enter a valid number.")

# Create the main window
window = tk.Tk()
window.title("Kilogram to Tons and Tons to Kilogram Converter")

# KG to Tons conversion
kg_label = tk.Label(window, text="Enter weight in kilograms:")
kg_label.pack()
kg_entry = tk.Entry(window)
kg_entry.pack()
kg_btn = tk.Button(window, text="Convert kg to tons", command=convert_kg_to_tons)
kg_btn.pack()
kg_result_label = tk.Label(window, text="")
kg_result_label.pack()

# Tons to KG conversion
tons_label = tk.Label(window, text="Enter weight in tons:")
tons_label.pack()
tons_entry = tk.Entry(window)
tons_entry.pack()
tons_btn = tk.Button(window, text="Convert tons to kg", command=convert_tons_to_kg)
tons_btn.pack()
tons_result_label = tk.Label(window, text="")
tons_result_label.pack()

# Run the Tkinter event loop
window.mainloop()
```

```python
#Find the sum and product of its digits.
import tkinter as tk
from tkinter import messagebox

def sum_and_product_of_digits(number):
    if 10 <= number <= 99:  # Ensure the number is two-digit
        tens = number // 10
        units = number % 10
        sum_digits = tens + units
        product_digits = tens * units
        return sum_digits, product_digits
    else:
        raise ValueError("The number must be a two-digit number")

def on_calculate():
    try:
        number = int(entry.get())
        sum_digits, product_digits = sum_and_product_of_digits(number)
        result_label.config(text=f"The sum of the digits is {sum_digits}.\nThe product of the digits is {product_digits}.")
    except ValueError as e:
        messagebox.showerror("Invalid Input", str(e))

# Create the main window
window = tk.Tk()
window.title("Sum and Product of Digits")

# Create and place the entry field
entry_label = tk.Label(window, text="Enter a two-digit number:")
entry_label.pack()
entry = tk.Entry(window)
entry.pack()

# Create and place the calculate button
calculate_button = tk.Button(window, text="Calculate", command=on_calculate)
calculate_button.pack()

# Create and place the result label
result_label = tk.Label(window, text="")
result_label.pack()

# Run the Tkinter event loop
window.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

# Predefined dictionary for translation
translation_dict = {
    "hello": "привет",
    "world": "мир",
    "cat": "кот",
    "dog": "собака",
    "good": "хорошо",
    "bad": "плохо",
    "morning": "утро",
    "evening": "вечер"
}

def translate():
    word = english_word.get().lower()
    translated_word = translation_dict.get(word, None)
    translation_box.config(state=NORMAL)
    translation_box.delete(1.0, END)
    if translated_word:
        translation_box.insert(END, translated_word)
        translation_box.config(state=DISABLED)
    else:
        translation_box.insert(END, "Translation not found. Please enter the translation and click Add.")

def clear_translation():
    english_word_entry.delete(0, END)
    translation_box.config(state=NORMAL)
    translation_box.delete(1.0, END)
    translation_box.config(state=DISABLED)

def add_translation():
    word = english_word.get().lower()
    translated_word = translation_box.get(1.0, END).strip()
    if word and translated_word and translated_word != "Translation not found. Please enter the translation and click Add.":
        translation_dict[word] = translated_word
        clear_translation()
        messagebox.showinfo("Success", f"Added translation: {word} -> {translated_word}")
    else:
        messagebox.showerror("Error", "Please enter a valid word and translation.")

root = Tk()
root.title("English to Russian Translator")
root.geometry("400x400")

# English word input
english_word_label = Label(root, text="Enter an English word:", font=("Arial", 14))
english_word_label.pack(pady=10)

english_word = StringVar()
english_word_entry = ttk.Entry(root, textvariable=english_word, font=("Arial", 14))
english_word_entry.pack(pady=5)

# Translate button
translate_button = ttk.Button(root, text="Translate", command=translate)
translate_button.pack(pady=10)

# Clear button
clear_button = ttk.Button(root, text="Clear", command=clear_translation)
clear_button.pack(pady=10)

# Add button
add_button = ttk.Button(root, text="Add", command=add_translation)
add_button.pack(pady=10)

# Translation result in a text box
translation_label = Label(root, text="Translation in Russian:", font=("Arial", 14))
translation_label.pack(pady=10)

translation_box = Text(root, height=1, width=30, font=("Arial", 14), state=NORMAL)
translation_box.pack(pady=5)
translation_box.config(state=DISABLED)

root.mainloop()
```
